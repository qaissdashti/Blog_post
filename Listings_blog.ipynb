<b> Change some of the object type cols into Numeric</b>

```python
# they are objects can be changed to numeric
for col in to_change_numeric:
    print(listing_df[col].dtype)
```

```python
# we use the below loop to remove the $ from the cols

for col in fixed_num.columns:
    fixed_num[col].replace({'\$':''}, regex=True, inplace=True)
```
